````python
# ============================ CELL 3 (APRIL 2026 FIX) =========================
# IMPORTANT: Gemini 2.0 Flash was deprecated in March 2026.
# Current working free-tier models: 2.5 Flash, 2.5 Flash-Lite, 2.5 Pro
#
# Copy this ENTIRE cell and REPLACE your Cell 3 in Colab

import base64
import io
import re
import time
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import google.generativeai as genai

# Configure Gemini
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# ============ CURRENT FREE TIER MODELS (April 2026) ============
# Will auto-fallback if one hits rate limit
MODELS = [
    "gemini-2.5-flash",          # 10 RPM, 250/day (best quality on free)
    "gemini-2.5-flash-lite",     # 15 RPM, 1000/day (highest throughput)
    "gemini-2.5-pro",            # 5 RPM, 100/day (slowest but most capable)
]

app = FastAPI(title="Sketch2Code AI")

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

class SketchRequest(BaseModel):
    image: str
    style: str = "modern"

SYSTEM_PROMPT = """You are an expert frontend developer. You receive a hand-drawn UI sketch (wireframe) and must convert it into production-ready HTML with Tailwind CSS classes.

RULES:
1. Output ONLY a single complete HTML file — nothing else. No explanations, no markdown fences.
2. Include Tailwind CDN: <script src="https://cdn.tailwindcss.com"></script>
3. Interpret the sketch GENEROUSLY — if you see a rough rectangle labeled "button", make it a beautiful button. Be creative in making it look polished.
4. Use realistic placeholder text (not "Lorem ipsum" — use real-sounding product copy).
5. Add nice shadows, rounded corners, proper spacing, good typography.
6. Use a cohesive color palette. Default to indigo/slate if unclear.
7. Make it RESPONSIVE (mobile-friendly).
8. Add smooth hover states on interactive elements.
9. Use Lucide icons via CDN if icons are needed.
10. The output must be a COMPLETE standalone HTML page starting with <!DOCTYPE html>."""

STYLE_NOTES = {
    "modern": "Use a modern, clean aesthetic with bold typography, generous whitespace, and a primary color of indigo-600. Add subtle gradients.",
    "minimal": "Use a strictly minimal aesthetic — black, white, one accent color. Thin borders, lots of whitespace, elegant serif headings.",
    "playful": "Use a playful aesthetic — rounded corners everywhere, pastel colors (pink-400, yellow-300, teal-400), fun emojis where appropriate.",
    "corporate": "Use a corporate professional aesthetic — navy blue primary, gray neutrals, sharp corners, serious tone, subtle shadows.",
}

def extract_html(text: str) -> str:
    text = text.strip()
    text = re.sub(r"^```(?:html)?\s*\n?", "", text)
    text = re.sub(r"\n?```\s*$", "", text)
    if not text.lower().lstrip().startswith("<!doctype") and not text.lower().lstrip().startswith("<html"):
        text = f"<!DOCTYPE html>\n<html><head><script src='https://cdn.tailwindcss.com'></script></head><body>{text}</body></html>"
    return text


def generate_with_fallback(prompt, pil_image):
    """Try multiple models — if one hits rate limit, try next."""
    last_error = None
    for model_name in MODELS:
        try:
            model = genai.GenerativeModel(model_name)
            response = model.generate_content([prompt, pil_image])
            print(f"✅ Success with: {model_name}")
            return response.text, model_name
        except Exception as e:
            error_str = str(e)
            print(f"⚠️ {model_name} failed: {error_str[:80]}")
            last_error = e
            if "429" in error_str or "quota" in error_str.lower():
                time.sleep(1)
            continue
    raise last_error or Exception("All models failed")


@app.get("/")
def root():
    return {"status": "ok", "service": "Sketch2Code AI", "version": "2.0", "models": MODELS}


@app.post("/generate")
def generate_code(req: SketchRequest):
    try:
        image_data = req.image
        if "," in image_data:
            image_data = image_data.split(",", 1)[1]
        img_bytes = base64.b64decode(image_data)

        from PIL import Image
        pil_image = Image.open(io.BytesIO(img_bytes))

        style_note = STYLE_NOTES.get(req.style, STYLE_NOTES["modern"])
        full_prompt = f"{SYSTEM_PROMPT}\n\nSTYLE: {style_note}\n\nNow convert this sketch into a beautiful HTML page:"

        raw_text, model_used = generate_with_fallback(full_prompt, pil_image)
        html = extract_html(raw_text)

        return {
            "success": True,
            "html": html,
            "style": req.style,
            "model": model_used,
        }

    except Exception as e:
        error_msg = str(e)[:300]
        if "429" in error_msg or "quota" in error_msg.lower():
            friendly = "All free models are rate-limited. Wait 1 minute or use a different Google account's API key."
        elif "API_KEY" in error_msg.upper() or "401" in error_msg:
            friendly = "Invalid Gemini API key. Check Cell 2."
        elif "404" in error_msg or "not found" in error_msg.lower():
            friendly = "Model not found. Google may have deprecated it — check latest model names at ai.google.dev"
        else:
            friendly = f"Error: {error_msg[:150]}"

        return {
            "success": False,
            "error": friendly,
            "html": f"<html><body style='font-family:sans-serif;padding:40px;background:#f3f4f6'><div style='max-width:600px;margin:auto;background:white;padding:32px;border-radius:16px;box-shadow:0 4px 20px rgba(0,0,0,0.1)'><h2 style='color:#dc2626;margin:0 0 16px'>⚠️ Error</h2><p style='color:#374151;line-height:1.6'>{friendly}</p></div></body></html>",
        }


print("✅ App ready (Gemini 2.5 models). Now run Cell 4.")
````
